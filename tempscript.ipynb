```python
# Parse all the lines that are like: Running the prover agent for lemma: mathd_numbertheory_320 with timeout: 334.60800099372864 seconds
# and extract the lemma name and the timeout
pattern = re.compile(r"Running the prover agent for lemma: (\w+) with timeout: (\d+(\.\d+)*) seconds")
matches = pattern.findall(evals)
time_per_lemma = {}
for thm_name, timeout, _ in matches:
    time_per_lemma[thm_name] = float(timeout)
print(len(time_per_lemma))
# for thm_name, timeout in time_per_lemma.items():
#     print(f"{thm_name}: {timeout}")
print(sum([v for v in time_per_lemma.values()]))
```

```python
# Get last execution time for all
# Parse something like
# LemmaName: theorem amc12b_2020_p2 :
# ((100 ^ 2 - 7 ^ 2):ℝ) / (70 ^ 2 - 11 ^ 2) * ((70 - 11) * (70 + 11) / ((100 - 7) * (100 + 7))) = 1 :=
# SearchResult: [SUCCESS]
# IsInferenceExhausted: False
# IsTimeout: False
# LongestSuccessPath: -1 
# StepsUsed: 1
# SearchTimeInSecs: 19.019124269485474
# NumberOfBacktracks: 0
# PossibleFailedPaths: -1
# AdditionalInfo: {'queries': 1}

#Extract LemmaName
# Multiline match till you find the first SearchTimeInSecs
pattern = re.compile(r"LemmaName: theorem (\w+)[\s\S]*?SearchTimeInSecs: (\d+(\.\d+)*)", re.DOTALL)
matches = pattern.findall(evals)
last_exec_time = {}
for thm_name, time, _ in matches:
    last_exec_time[thm_name] = float(time)
print(len(last_exec_time))
print(sum([v for v in last_exec_time.values()]))
```

```python
# Remove the last execution time from the time_per_lemma
print(sum([v for v in time_per_lemma.values() if v > 0.0]))
for thm_name in all_thms:
    if thm_name in time_per_lemma and thm_name in last_exec_time:
        time_per_lemma[thm_name] -= last_exec_time[thm_name]
    else:
        time_per_lemma[thm_name] = 0
print(len(time_per_lemma))
print(sum([v for v in time_per_lemma.values() if v > 0.0]))
```

```python
# dump the time_per_lemma to a json file
for thm_name in all_thms:
    timeout = time_per_lemma[thm_name]
    print(f"{thm_name}: {timeout}")
```
